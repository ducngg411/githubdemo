```python
f = open('languages.txt', 'r') # open file with read mode
contents = f.read() #read all contents into a string
# print(contents)
languages = contents.split('\n')
print(len(languages))
for lan in languages:
    print(lan)
f.close() # close file
```

```python
f = open('languages.txt' , 'r')
lines = f.readlines() # read all contents into a list of strings
f.close()

for languages in lines:
    print(f'|{languages.strip()}') #strip() removes the newline character
```

```python
# Use this method for large file 
f = open('languages.txt', 'r')
line = f.readline() #read 1st line

while line != '':
    print(line.strip())
    line = f.readline() # read the next line
f.close()
```

```python
f = open('agencies.txt', 'r')
find_word = f.readline()

while find_word != '':
    if 'Authority' in find_word:
        print(find_word.strip())
    find_word = f.readline()
f.close()
```

```python
invalid_file  = True
while invalid_file:
    try:
        file_name = input('Enter file name: ')
        f = open(file_name, 'r')
        invalid_file = False
    except FileNotFoundError:
        print(f'File {file_name} not found. Try again.')

lines = f.readlines()
f.close()

f = open(file_name + '.authority', 'w')

for i, line in enumerate(lines):
    if 'Authority' in line:
        print(i+1, line.strip())
        f.write(line)

f.close()
```

```python
# Exercise: read authors.txt, find all authors that use gmail and save them 
# into a new file gmail_authors.txt in following format:
# Paul Bakaus: paul.bakaus@gmail.com

try: 
    f = open('authors.txt', 'r')
    lines = f.readlines()
    f.close()

except FileNotFoundError:
    print(f'File Not Found')

if lines:
    with open('gmail_authors.txt', 'w') as f:
        for line in lines:
            if 'gmail' in line:
                author = line.split('<')[0].strip()
                email = line.split('<')[1].strip().strip('>')
                f.write(f'{author}: {email} \n')
```

```python
import os

def read_folder(path):
    print(path)
    if os.path.isdir(path):
        list_paths = os.listdir(path)
        for p in list_paths:
            read_folder(path + '/' + p)

read_folder('D:\CitiesSkylines')
```

```python
def read_folder(path, level):
    print_spaces(level)
    if os.path.isdir(path):
        print('+', path.split('/')[-1])
    else:
        print('-', path.split('/')[-1])
    if os.path.isdir(path):
        level += 1
        list_paths = os.listdir(path)
        for p in list_paths:
            read_folder(path + '/' + p, level)

def print_spaces(level):
    for i in range(level):
        print(' ', end = '')

read_folder()
```
